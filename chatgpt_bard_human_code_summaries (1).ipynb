```python
# Install necessary libraries
# You can install these using pip if you haven't already
# pip install textstat nltk

import textstat
import nltk
import re
from nltk.tokenize import sent_tokenize, word_tokenize

# Download NLTK data (if you haven't already)
nltk.download('punkt')

# Define functions to calculate the Flesch-Kincaid scores using textstat
def flesch_kincaid_scores(text):
    # Calculate Flesch Reading Ease
    reading_ease = textstat.flesch_reading_ease(text)

    # Calculate Flesch-Kincaid Grade Level
    grade_level = textstat.flesch_kincaid_grade(text)

    return reading_ease, grade_level

# Function to count syllables in a word
def count_syllables(word):
    word = word.lower()
    syllables = 0
    vowels = "aeiouy"
    if word[0] in vowels:
        syllables += 1
    for index in range(1, len(word)):
        if word[index] in vowels and word[index - 1] not in vowels:
            syllables += 1
    if word.endswith("e"):
        syllables -= 1
    if word.endswith("le") and len(word) > 2 and word[-3] not in vowels:
        syllables += 1
    if syllables == 0:
        syllables += 1
    return syllables

# Function to get text statistics
def text_statistics(text):
    sentences = sent_tokenize(text)
    words = word_tokenize(text)
    syllables = sum(count_syllables(word) for word in words)

    num_sentences = len(sentences)
    num_words = len(words)
    num_syllables = syllables

    return num_sentences, num_words, num_syllables

# Function to calculate Flesch-Kincaid scores manually
def flesch_kincaid_manual(text):
    num_sentences, num_words, num_syllables = text_statistics(text)

    # Flesch Reading Ease
    reading_ease = 206.835 - 1.015 * (num_words / num_sentences) - 84.6 * (num_syllables / num_words)

    # Flesch-Kincaid Grade Level
    grade_level = 0.39 * (num_words / num_sentences) + 11.8 * (num_syllables / num_words) - 15.59

    return reading_ease, grade_level

# Sample text
text = "The code snippet initializes two variables, x and y, with the values 5 and 10, respectively. It then calculates the sum of these two variables and prints the result. The purpose of this code is to demonstrate a simple addition operation and output the sum, which in this case is 15."

# Using textstat
reading_ease, grade_level = flesch_kincaid_scores(text)
print(f"Flesch Reading Ease (textstat): {reading_ease}")
print(f"Flesch-Kincaid Grade Level (textstat): {grade_level}")

# Using manual calculation
reading_ease_manual, grade_level_manual = flesch_kincaid_manual(text)
print(f"Flesch Reading Ease (Manual): {reading_ease_manual}")
print(f"Flesch-Kincaid Grade Level (Manual): {grade_level_manual}")
```

```python
import time
import resource

def prime(x, y):
    prime_list = []
    for i in range(x, y):
        if i == 0 or i == 1:
            continue
        else:
            is_prime = True
            for j in range(2, int(i**0.5) + 1):
                if i % j == 0:
                    is_prime = False
                    break
            if is_prime:
                prime_list.append(i)
    return prime_list

# Function to get memory usage
def get_memory_usage():
    usage = resource.getrusage(resource.RUSAGE_SELF).ru_maxrss
    # Convert to megabytes (divide by 1024*1024)
    return usage / 1024.0

# Driver program
starting_range = 2
ending_range = 100000

# Record start time
start_time = time.time()

lst = prime(starting_range, ending_range)

# Record end time
end_time = time.time()

# Calculate the time taken
execution_time = end_time - start_time

# Get memory usage
memory_usage = get_memory_usage()

if len(lst) == 0:
    print("There are no prime numbers in this range")
else:
    print("The prime numbers in this range are: ", lst[:10], "... (total", len(lst), "primes)")

print(f"Time taken to find prime numbers: {execution_time:.6f} seconds")
print(f"Memory usage: {memory_usage:.2f} MB")
```

```python
import time
import resource

def sieve_of_eratosthenes(start, end):
    sieve = [True] * (end + 1)
    sieve[0] = sieve[1] = False  # 0 and 1 are not prime numbers

    for i in range(2, int(end**0.5) + 1):
        if sieve[i]:
            for j in range(i*i, end + 1, i):
                sieve[j] = False

    prime_list = [i for i in range(start, end + 1) if sieve[i]]
    return prime_list

# Function to get memory usage in KB (works on macOS)
def get_memory_usage():
    return resource.getrusage(resource.RUSAGE_SELF).ru_maxrss / 1024

# Driver program
starting_range = 2
ending_range = 100000

# Record start time
start_time = time.time()

# Get initial memory usage
start_memory_usage = get_memory_usage()

lst = sieve_of_eratosthenes(starting_range, ending_range)

# Get end time and memory usage
end_time = time.time()
end_memory_usage = get_memory_usage()

# Calculate the time taken
execution_time = end_time - start_time

if len(lst) == 0:
    print("There are no prime numbers in this range")
else:
    print("The prime numbers in this range are: ", lst)

print(f"Time taken to find prime numbers: {execution_time} seconds")
print(f"Memory usage: {end_memory_usage - start_memory_usage:.2f} KB")
```

```python
import nltk
import numpy as np
from scipy.stats import ttest_rel
from nltk.tokenize import word_tokenize, sent_tokenize
from textstat import flesch_reading_ease

# Download necessary NLTK data files
nltk.download('punkt')

# Text data
texts = {
    'bard_add': (
        "This code snippet performs a simple addition and prints the result."
        "It first assigns the value 5 to variable x and 10 to variable y."
        "Then, it uses the print function to display the sum of x and y on the console."
        "In essence, the code adds two numbers and shows their combined value."
    ),
    'human_add': (
        "The given python program performs addition."
        " The code provided initializes two variables, x with a value of 5 and y with a value of 10, and then prints the sum of these two variables."
    ),
    'bard_euc': (
        "This code defines a function gcd(a, b) that calculates the greatest common divisor (GCD) of two numbers a and b."
        "It implements the Euclidean algorithm. Here's how it works:"
        "1.	Base Case: If a is zero, the GCD is simply b and the function returns b."
        "2.	Recursive Case: Otherwise, the function calculates the remainder (b % a) when b is divided by a."
        "It then makes a recursive call to gcd with b % a (remainder) as the new a and the original a as the new b."
        "The code then uses the gcd function three times with different values for a and b, printing the GCD for each pair."

    ),
    'human_euc': (
        "In this code snippet, we are implementing a function to calculate the Greatest Common Divisor (GCD) of two numbers using the Euclidean algorithm."
        "The GCD is the largest positive integer that divides each of the numbers without leaving a remainder."
        "The code defines a function gcd(a, b) that calculates the GCD of two numbers a and b using the Euclidean algorithm."
        "It then demonstrates the function by calculating the GCD of different pairs of numbers."
    ),
    'bard_lrg': (
        "This code defines a function largest(arr, n) that finds the largest element in an array arr of size n. Here's how it works:"
        "1.	It initializes a variable max with the first element of the array, assuming it's the largest initially."
        "2.	It iterates through the array using a loop, starting from the second element (index 1) up to n-1 (to cover all elements)."
        "3.	Inside the loop, it compares each element arr[i] with the current max."
        "4.	If the current element is greater than max, it updates max to hold the value of the larger element."
        "5.	After iterating through the entire array, the function returns the final value of max, which is guaranteed to be the largest element in the array."
        "The code then creates an array arr with some sample values and calls the largest function with arr and its length n."
        "Finally, it prints the largest element found in the array."

    ),
    'human_lrg': (
        "In this code snippet, we have a Python function that finds the largest element in an array."
        "The function takes an array and its length as input and returns the largest element in the array."
        "The code consists of a function largest that finds the largest element in an array and a simple script that demonstrates the function by finding the largest element in a predefined array."
    ),
    'bard_read': (
        "This code snippet reads the contents of a text file and prints each word on a separate line. Here's how it works:"
        "1.	It uses the with open statement to open a file named 'GFG.txt' in read mode ('r')."
        "The with statement ensures the file gets closed automatically after the code within the indented block finishes."
        "2.	It iterates over each line in the opened file using a for loop."
        "3.	Inside the inner loop, it splits the current line using the split() method."
        "This splits the line into a list of words based on whitespace (spaces, tabs, etc.) by default."
        "4.	Another for loop iterates over the words obtained from splitting the line."
        "5.	Within the inner loop, it simply prints each individual word on a separate line."
        "In essence, the code reads the file line by line, separates the words in each line, and then prints each word independently."


    ),
    'human_read': (
        "In this code snippet, we are reading a text file named 'GFG.txt' in Python using open() function and then iterating through each line of the file using nested for loops."
        "For each line, we further split the line into individual words using split() function and print each word on a new line."
    ),
    'bard_obj': (
        "This code defines a class named GFG that represents an employee with a name and company. Here's a breakdown of how it works:"
        "1.	Class Definition: The code defines a class GFG with two attributes: name and company."
        "2.	Constructor: It also defines a special method __init__, which acts as the constructor for the class."
        "This method is called whenever a new object (instance) of the class is created."
        "It takes two arguments, self (referring to the object itself) and two additional arguments (name and company) that represent the employee's details. Inside the constructor, it assigns the provided values to the object's attributes (self.name and self.company)."
        "3.	Instance Method: The class also defines another method named show."
        "This method doesn't take any arguments because it operates on the data of the object itself."
        "It uses string concatenation to create a message combining the object's name, company, and then prints it using the print function."
        "4.	Object Creation and Method Call: After defining the class, the code creates an instance of GFG named obj by providing 'John' as the name and 'GeeksForGeeks' as the company."
        "Finally, it calls the show method on the obj instance, which prints the greeting message with the specific details provided during object creation."
        "In essence, this code demonstrates creating a class to represent an employee and then using that class to create objects (instances) with specific employee data and printing information about them."

    ),
    'human_obj': (
        "In this code snippet, we have a Python class named GFG that demonstrates the concept of class initialization and object creation."
        "The code consists of a class GFG with an __init__ method to initialize the object with name and company attributes."
        "It also includes a show method to print a message containing the object's name and company."
        "Finally, an object obj of class GFG is created with the name 'John' and company 'GeeksForGeeks', and the show method is called on this object."
    )
}

# Function to extract features from text
def extract_features(text):
    words = word_tokenize(text)
    sentences = sent_tokenize(text)
    word_count = len(words)
    avg_word_length = np.mean([len(word) for word in words])
    sentence_count = len(sentences)
    avg_sentence_length = np.mean([len(sentence.split()) for sentence in sentences])
    readability = flesch_reading_ease(text)
    return [word_count, avg_word_length, sentence_count, avg_sentence_length, readability]

# Extract features for all texts
features = {key: extract_features(text) for key, text in texts.items()}

# Perform paired t-tests for each pair of texts
results = {}
pairs = [
    ('bard_add', 'human_add'),
    ('bard_euc', 'human_euc'),
    ('bard_lrg', 'human_lrg'),
    ('bard_read', 'human_read'),
    ('bard_obj', 'human_obj')
]

for chatgpt_key, human_key in pairs:
    t_stat, p_value = ttest_rel(features[chatgpt_key], features[human_key])
    results[(chatgpt_key, human_key)] = {
        'features_bard': features[chatgpt_key],
        'features_human': features[human_key],
        't_stat': t_stat,
        'p_value': p_value
    }

# Print results
for pair, result in results.items():
    chatgpt_key, human_key = pair
    print(f"Comparison between {chatgpt_key} and {human_key}:")
    print(f"Features (Bard): {result['features_bard']}")
    print(f"Features (Human): {result['features_human']}")
    print(f"T-statistic: {result['t_stat']}")
    print(f"P-value: {result['p_value']}")
    print()
```
